# NumPy
科学技術計算用ライブラリ

$n$次元配列や数学の演算に頻繁に使用

特に配列の記述法は今後の研究にも大いに役立つはず


```python
# NumPyによる行列の記述法

import numpy as np # 通常npという略称で取り扱う

x = np.array([[1, 2, 3], [4, 5, 6]])
print("x:\n{}".format(x))
```

# SciPy

こちらも科学技術計算用ライブラリ

特に数学関数や統計処理に強い

```python
# SciPyによる疎行列の定義

from scipy import sparse # ライブラリ中の1つの関数を使いたいときはfromを使用

# 4次の単位行列
eye = np.eye(4) # 上のセルでNumPyを呼び出しているのでここでは不要
print("NumPy array:\n{}".format(eye))

sparse_matrix = sparse.csr_matrix(eye)
print("\nSciPy sparse CSR matrix:\n{}".format(sparse_matrix))
```

# Matplotlib

グラフ描画用ライブラリ

通常の折れ線グラフや棒グラフの他，ヒストグラムや箱ひげ図もライブラリに含まれる

# Pandas

RのデータフレームをPythonで模して扱うライブラリ

ビッグデータ解析時には必須

Pandasもまたクセが強め

```python
import pandas as pd # 通常pdという略名で使用

# 以下のデータ構造はPythonだとディクショナリと呼ばれる
# Java等では連想配列と呼ばれるもの
data = {
    'Name': ["John", "Anna", "Peter", "Linda"],
    'Location': ["New York", "Paris", "Berlin", "London"],
    'Age': [24, 13, 53, 33]
}

# dataをデータフレームに格納
data_pandas = pd.DataFrame(data)

display(data_pandas)

# ageが30以上のすべての行を抽出

display(data_pandas[data_pandas.Age >= 30])
```

# Scikit-Learn

Pythonの機械学習用ライブラリ

頻繁にバージョンアップされている模様

```python
from sklearn.datasets import load_iris # 超有名なアイリスの種類識別用データセット

iris_dataset = load_iris()
print("Key of iris_dataset: \n{}".format(iris_dataset.keys()))
```

```python
print(iris_dataset['DESCR'][:193] + '\n...') # ":n"という表記は最初からn-1番目までを表す
```

```python
# 識別用ラベル
print("Target names: {}".format(iris_dataset['target_names']))
```

```python
# 予測に用いる入力データの種類
print("Feature names: \n{}".format(iris_dataset['feature_names']))
```

```python
# 入力データのデータ型
print("Type of data: {}".format(type(iris_dataset['data'])))
```

```python
# データサイズ (ここが超重要)
print("Shape of data: {}".format(iris_dataset['data'].shape))
```

```python
# データの中身 (5件)
print("First 5 columns of data:\n{}".format(iris_dataset['data'][:5]))
```

```python
# ラベルのタイプ
print("Type of target: {}".format(type(iris_dataset['target'])))

# ラベルのデータサイズ
print("Shape of target: {}".format(iris_dataset['target'].shape))

# 実際のデータ
print("Target:\n{}".format(iris_dataset['target']))
```

```python
# データセットを訓練データと検証データに分割
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)
# 変数を連ねて記述して各変数に代入することも可能

# 訓練データセットのサイズ
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))

# 検証データセットのサイズ
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
```

```python
# データの観察

iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names) # クラス的な表記も可能

grr = pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15), marker='o', s=60, alpha=0.8)
```

```python
# k-NNの実行

from sklearn.neighbors import KNeighborsClassifier

# k-NNオブジェクト作成
knn = KNeighborsClassifier(n_neighbors=1)

# 予測モデル作成
knn.fit(X_train, y_train)
```

```python
# モデルの評価

# まずは検証データセットを用いて予測
y_pred = knn.predict(X_test)
print("Test set predictions:\n{}".format(y_pred))

# 本来のラベルとの一致度を評価
print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
```
